## Historical Stock Price Data Cleansing

```python
# Read in data

stock_df = pd.read_csv("/Users/johngalvin/Desktop/GitHub/Project2_Fram_Galvin_Bradford/data/raw/HistoricalStockPriceData.csv")
```

```python
# Examine

stock_df.head()
```

```python
# Drop unnecessary columns

stock_df.drop(["Class", "Full BBG Ticker", "Date Show Flag", "Start Date", "End Date", "Unnamed: 6"], axis=1, inplace=True)
```

```python
# Transpose the table (swap columns and rows)

stock_df = stock_df.T
```

```python
# Set columns as first row (ticker labels)

header = stock_df.iloc[0]
stock_df.columns = header
stock_df.rename(columns=stock_df.iloc[0])
stock_df = stock_df.iloc[1:]
```

```python
# Remove 10/1/2018 record as all are undefined

stock_df = stock_df.iloc[1:]
```

```python
# Make the index datetime

stock_df.index = pd.to_datetime(stock_df.index)
```

```python
# Remove all "," from cells

stock_df = stock_df.replace({",": ""}, regex=True)
```

```python
# Recast all str values as float

stock_df = stock_df.astype(float)
```

```python
# Count NaN values

stock_df.isnull().sum().sum()
```

```python
# Impute NaN values with column-wise median
# Opt for not using KNN because that requires distance-based metrics and standardization of all values

stock_df.fillna(stock_df.median, axis=1, inplace=True)
```

```python
# Rename index

stock_df.index.rename("Date", inplace=True)
```

```python
# Write dataframe out to csv

stock_df.to_csv("/Users/johngalvin/Desktop/GitHub/Project2_Fram_Galvin_Bradford/data/processed/historical_stock_clean.csv", index=True)
```

## Insider Transaction Data Cleansing

```python
# Read in data

insider_df = pd.read_csv("/Users/johngalvin/Desktop/GitHub/Project2_Fram_Galvin_Bradford/data/raw/Insidertxns.csv")
```

```python
# Examine

insider_df.head()
```

```python
# Drop unnecessary columns

insider_df.drop(["Exchange", "Currency", "PE Ratio", "Cost(000)", "Final Share", "Market Cap ($M)"], axis=1, inplace=True)
```

```python
# Make the index datetime and get rid of extra Date column

insider_df.index = pd.to_datetime(insider_df["Date"])
insider_df.drop("Date", axis=1, inplace=True)
```

```python
insider_df.head()
```

```python
# Recast all appropriate columns as float or int

insider_df["Price"] = insider_df["Price"].astype(float)
insider_df["Insider Trading Shares"] = insider_df["Price"].astype(int)
insider_df["Shares Change"] = insider_df["Shares Change"].astype(float)
insider_df["Price.1"] = insider_df["Price.1"].astype(float)
insider_df["Price Change Since Insider Trade (%)"] = insider_df["Price Change Since Insider Trade (%)"].astype(float)
insider_df["Dividend Yield %"] = insider_df["Dividend Yield %"].astype(float)
```
